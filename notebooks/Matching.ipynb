## Graph Matching

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

from graspy.match import GraphMatch as GMP
from graspy.simulations import er_corr, sbm, sbm_corr
from graspy.plot import heatmap
%matplotlib inline
```

```python
n = 50
p = 0.3

G1 = er_np(n, p)
node_shuffle_input = np.random.permutation(n)
G2 = G1[np.ix_(node_shuffle_input, node_shuffle_input)]

sum(sum(abs(G1-G2)))
```

```python
n = 50
p = .3
r = .7

G1, G2 = er_corr(n, p, r)

fix, axes = plt.subplots(1, 2, figsize=(10,5))
heatmap(G1, ax=axes[0])
heatmap(G2, ax=axes[1])
```

```python
gmp = GMP()
gmp.fit(G1, G2)
```

## Seeded graph matching

```python
n = [100, 100, 100]
p = np.array([[.7, .3, .4], 
              [.3, .7, .3], 
              [.4, .3, .7]])
r = .9

G1, G2 = sbm_corr(n, p, r)

heatmap(p, annot=True, cbar=False)
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
heatmap(G1, ax=axes[0], cbar=False, title="G1")
heatmap(G2, ax=axes[1], cbar=False, title="G2")
heatmap(G1-G2, ax=axes[2], cbar=False, title="G1-G2");
```
