# Plotting

```python
import graspy
import numpy as np
%matplotlib inline
```

```python
from graspy.simulations import sbm

P = np.array([[.8, .2],
              [.2, .8]])
n_communities = [50, 50]

# Draw weights from a poisson distribution
weight = np.random.poisson
wtargs = dict(lam=3)
G_poisson = sbm(n_communities, P, wt=weight, wtargs=wtargs)

# Draw weights from a normal distribution
wt = np.random.normal
wtargs = dict(loc=5, scale=1)
G_normal = sbm(n_communities, P, wt=wt, wtargs=wtargs)
```

```python
from graspy.plot import heatmap

heatmap(G_poisson)
heatmap(G_normal)
```

```python
fig = heatmap(G_poisson, cmap="binary")
```
